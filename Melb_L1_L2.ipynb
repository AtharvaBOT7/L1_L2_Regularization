```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('melb_data.csv')
```

```python
dataset.head(10)
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
cols = ['Suburb','Address','Rooms','Type','Price','Method','SellerG','Distance','Postcode','Bedroom2','Bathroom','Car','YearBuilt',
        'BuildingArea','Propertycount']

dataset = dataset[cols]
```

```python
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_zero = ['Car','YearBuilt']

dataset[cols_zero] = dataset[cols_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head(10)
```

```python
dataset.shape
```

```python
X = dataset.drop('Price', axis = 1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train_X,test_X, train_y,test_y = train_test_split(X,y,test_size=0.3,random_state = 42)
```

```python
reg = LinearRegression().fit(train_X,train_y)
```

```python
reg.score(train_X,train_y)
```

```python
reg.score(test_X,test_y)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(train_X,train_y)
```

```python
lasso_reg.score(train_X,train_y)
```

```python
lasso_reg.score(test_X,test_y)
```

```python
from sklearn.linear_model import Ridge
```

```python
rig = Ridge(alpha=50,max_iter =100, tol=0.1)
rig.fit(train_X,train_y)
```

```python
rig.score(train_X,train_y)
```

```python
rig.score(test_X,test_y)
```
